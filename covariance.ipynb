```python
import numpy as np
```

```python
x=np.array([1,3,5,7,9])
y=np.array([2,4,6,8,10])
```

```python
x_mean=np.mean(x)
y_mean=np.mean(y)
```

```python
covariance = np.sum((x - x_mean) * (y - y_mean)) / (len(x) - 1)
```
