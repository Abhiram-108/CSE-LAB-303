```python
import numpy as np
import pandas as pd
df=pd.read_csv('/content/Enjoy sports (1).csv')
df.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
x=df.drop('Decision',axis=1)
y=df['Decision']
x
```

```python
x['Outlook']=le.fit_transform(x['Outlook'])
x['Temp.']=le.fit_transform(x['Temp.'])
x['Humidity']=le.fit_transform(x['Humidity'])
x['Wind']=le.fit_transform(x['Wind'])
y=le.fit_transform(y)
```

```python
x=x.drop('Day',axis=1)
```

```python
x
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=0
)
```

```python
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    ConfusionMatrixDisplay,
    f1_score,
)
model.fit(x_train,y_train)

y_pred = model.predict(x_test)
accuray = accuracy_score(y_pred, y_test)
f1 = f1_score(y_pred, y_test, average="weighted")

print("Accuracy:", accuray)
print("F1 Score:", f1)
```

```python
conf_mat = confusion_matrix(
    y_test, y_pred, labels=[0, 1, 2] )

print(accuray)
print(conf_mat)
```
