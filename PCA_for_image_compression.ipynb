```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from skimage import data
from skimage.color import rgb2gray
import numpy as np

```

```python
grayscale_image = rgb2gray(data.astronaut())
```

```python
grayscale_image = grayscale_image[:256, :256]
```

```python
image_flattened = grayscale_image.reshape(grayscale_image.shape[0], -1)
```

```python
def reconstruct_with_pca(pca_model, num_components, image_data):
    """Reconstruct the image using the first num_components principal components."""
    pca_model.n_components = num_components
    pca_transformed = pca_model.fit_transform(image_data)
    image_reconstructed = pca_model.inverse_transform(pca_transformed)
    return image_reconstructed
```

```python
pca_model = PCA()
```

```python
component_counts = [5, 20, 50, 100]
```

```python
plt.figure(figsize=(12, 8))
```

```python
plt.subplot(2, 3, 1)
plt.imshow(grayscale_image, cmap='gray')
plt.title('Original Image (256x256)')
plt.axis('off')
```

```python
for index, num_components in enumerate(component_counts, 2):
    image_reconstructed = reconstruct_with_pca(pca_model, num_components, image_flattened)
    plt.subplot(2, 3, index)
    plt.imshow(image_reconstructed, cmap='gray')
    plt.title(f'{num_components} Principal Components')
    plt.axis('off')
```

```python

plt.tight_layout()
plt.show()
```
